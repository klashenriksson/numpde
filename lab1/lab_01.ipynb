```python
# Arrary and stuff 
import numpy as np
# Linear algebra solvers from scipy
import scipy.linalg as la
# Basic plotting routines from the matplotlib library 
import matplotlib.pyplot as plt
```

Next we define the grid points.

```python
# Number of equally spaced subintervals
N = 4
# Mesh size
h = 1/N #Important! In Python 2 you needed to write 1.0 to prevent integer divsion
# Define N+1 grid points via linspace which is part of numpy now aliased as np 
x = np.linspace(0,1,N+1)
print(x)
```

Now define matrix $A$ and right-hand side vector $F$.
We will first fill in the values that will be unchanged for different
boundary conditions.

```python
# Define a (full) matrix filled with 0s.
A = np.zeros((N+1, N+1))

# Define tridiagonal part of A by for rows 1 to N-1
for i in range(1, N):
    A[i, i-1] = -1
    A[i, i+1] = -1
    A[i, i] = 2
    
# Define right hand side. Instead of iterating we
# use a vectorized variant to evaluate f on all grid points
# Look out for the right h factors! 
F = h**2*(2*np.pi)**2*np.cos(2*np.pi*x)

# Note that F[0] and F[N] are also filled!
```

Last step to set up the system is to take the boundary conditions into account by modifying $A$ and $F$ properly.

```python
# Left boundary
A[0,0] = 1
F[0] = 1

# Right boundary
A[N,N] = 1
F[N] = 1
```

```python
def difference(f, boundry_condition):
    N = len(f)-1
    h = 1/N
    A = np.zeros((N+1, N+1))
    for i in range(1, N):
        A[i, i-1] = -1
        A[i, i+1] = -1
        A[i, i] = 2

    F = f*h**2
    
    if boundry_condition[0] == 'dirchlet':
        A[0,0] = 1
        A[N,N] = 1
        F[0] = boundry_condition[1]
        F[N] = boundry_condition[2]
    elif boundry_condition[0] == 'neuman_dirchlet':
        A[0,0] = -1
        A[0,1] = 1
        A[N,N] = 1
        F[0] = boundry_condition[1]*h
        F[N] = boundry_condition[2]
    elif boundry_condition[0] == 'neuman':
        A[0,0] = -1
        A[0,1] = 1
        A[N,N] = 1
        A[N, N-1] = -1
        F[0] = boundry_condition[1]*h
        F[N] = boundry_condition[2]*h
    elif boundry_condition[0] == 'ghost':
        A[0,0] = 1
        A[0,1] = -1
        A[N,N] = 1
        F[0] = h**2*f[0]*0.5 - h*boundry_condition[1]
        F[N] = boundry_condition[2]

    U = la.solve(A, F)
    return U
```

Now we solve the linear algebra system $AU = F$ and plot the results.

```python
Ns = [4,8,16,32,64]
markers = ['xr--', 'xg--', 'xb--', 'xk--', 'xc--']
for N, marker in zip(Ns, markers):
    x = np.linspace(0,1,N+1)
    f = (2*np.pi)**2*np.cos(2*np.pi*x)

    U_d = difference(f, ("dirchlet", 1, 1))
    U_nd = difference(f, ("neuman_dirchlet", 0, 1))
    #U_nn = difference(f, ("neuman", 0, 0))

    plt.figure(0)
    plt.plot(x,U_d, marker, label=f'N={N}')
    plt.figure(1)
    plt.plot(x,U_nd, marker,label=f'N={N}')

# analytic
x = np.linspace(0,1,100)
U_analytic = np.cos(2*np.pi*x)
plt.figure(0)
plt.plot(x,U_analytic, '-g')
plt.legend()
plt.figure(1)
plt.plot(x,U_analytic,'-g')
plt.legend()
plt.show()
    
```

## Problem 4

The goal of this problem is to investigate the numerical error introduced by the FDM more quantitatively and to familiarize us with the **method of manufactured solution**.

The idea is to assess the accurracy and correctness of a PDE solver implementation by constructing a know reference solution which solves the PDE problem at hand. This can be simply done by picking a meaningful and not too boring analytical solution
and explicitly calculate the data which need to be supplied, e.g.,
the right-hand side or boundary values for various boundary condition.

For instance, taking the function $u(x) = x + \cos(2 \pi x)$, we can simply calculate that
$$
\begin{align*}
 u'(x) &= 1 - 2\pi \sin(2\pi x)
 \\
 u''(x) &= - (2\pi)^2 \cos(2\pi x)
\end{align*}
$$
and thus $u$ satisfies the Poisson problem
$$
-u''(x) = (2\pi)^2 \cos(2\pi x)
$$
with boundary conditions
$$
u'(0) = 1, \quad u(1) = 2.
$$

With a known reference solution at hand we can compute
the error vector $u_i - U_i$ at the grid points $\{x_i\}_{i=0}^N$
for a series of successively refined grid, e.g. by taking
$N = 4\cdot2^k$ for $k = 0, 1, 2, 3$ etc.

Reducing the mesh size $h$ by half allows us to easily compute
the **experimental order of convergence (EOC)**,
that is the observed error reduction in the numerical solution
when passing from a coarser mesh with mesh size $h$ to
to a finer mesh with $h/2$.
The EOC can then be compare with the theoretically predicted error reduction (if known).

For instance, if you know that the discretization error $E(h)$ given on mesh
with mesh size $h$ and measured in some norm $ \| \cdot \| $ behaves like 
$\| E(h) \| \sim h^k$, you can conclude that  
$$
\dfrac{\|E(h/2)\|}{\|E(h)\|} \sim \dfrac{(h/2)^k}{h^k} = (1/2)^k
$$
when passing from $h$ to half the mesh size $h/2$. Taking the logarithm
of the last equation shows 
$$
k \sim \dfrac{\ln\bigl(\|E(h)\|/\|E(h/2)\|\bigr)}{\ln 2}. 
$$
(Verify this!)
Thus the EOC for each mesh size h is measured by the 
$$
EOC(h) = \dfrac{\ln\bigl(\|E(h)\|/\|E(h/2)\|\bigr)}{\ln 2}.
$$

Alternatively, you can do a log-log plot of your error as function of $h$,
that is plot $\ln (\| E(h)\|)$ againts $\ln(h)$. Then the slope of this
plot should be $\sim k$ if we expect the method to be of convergence order $k$.

# Answers

## a)

We have
$$
\begin{align*}
  u(x) = x - \dfrac{e^{(x-1)/\epsilon} - e^{-1/\epsilon}}{1 - e^{-1/\epsilon}}
\end{align*}
$$
Differentiating yields
$$
\frac{du}{dx} = 1 - \frac{1}{\epsilon}\frac{e^{(x-1)/\epsilon}}{1 - e^{-1/\epsilon}}
$$
Differentiating again yields
$$
\frac{d^2u}{dx^2} = -\frac{1}{\epsilon^2}\frac{e^{(x-1)/\epsilon}}{1 - e^{-1/\epsilon}}
$$

Inserting the above expressions for $\frac{d^2u}{dx^2}$ and $\frac{du}{dx}$ into the *Advection-Diffusion* PDE gives an expression for $f(x)$:
$$
-\epsilon\left(-\frac{1}{\epsilon^2}\frac{e^{(x-1)/\epsilon}}{1 - e^{-1/\epsilon}}\right) + b\left(1 - \frac{1}{\epsilon}\frac{e^{(x-1)/\epsilon}}{1 - e^{-1/\epsilon}}\right) = f(x)
$$
For $b = 1$ this yields
$$
f(x) = 1
$$
## b)
Replacing $u''$ and $u'$ with their respective numerical approximations (FDM stencils) yields the following expression (here we have used the center stencil for both $u''$ and $u'$):
$$
\epsilon\frac{-u(x-h) + 2u(x) - u(x+h)}{h^2} + \frac{u(x-h) + u(x+h)}{2h} = 1
$$
By discretizing the domain into a mesh of N points with an equal spacing of $h$ between each point, we can evaluate the above expression in each mesh point $x_i$ and get the following expression:
$$
\epsilon\frac{-u(x_{i-1}) + 2u(x_i) - u(x_{i+1})}{h^2} + \frac{u(x_{i-1}) + u(x_{i+1})}{2h} = 1
\Leftrightarrow
-u(x_{i-1}) + 2u(x_i) - u(x_{i+1}) + h\frac{u(x_{i-1}) + u(x_{i+1})}{2\epsilon} = \frac{h^2}{\epsilon}
$$
As this can be done for each $x_i$ we get a system of linear equations of the form
$$
\mathbf{A}\vec{u} = \vec{\frac{h^2}{\epsilon}}.
$$
See code below for the exact matrix.

## c)
We see from the figures below that the FDM solution becomes unstable as $\epsilon$ decreases. The approximation also have a distinct zig-zag pattern around the analytical solution.

The EOC for the different $\epsilon$s is given below. From the tables we see that for large values of $\epsilon$s, the order of convergence seem to be $2$ for all mesh sizes tested. However, as $\epsilon$ decreases, this is no longer the case. For $\epsilon = 0.01$ the order of convergence is much lower than $2$ for the smaller mesh sizes, but seem to approach $2$ for larger mesh sizes. For $\epsilon = 0.001$, the order of convergence does not seem to approach $2$ for any of the mesh sizes, suggesting an even finer mesh is required.

## d)

From the figures below we see that for the larger values of $\epsilon$, the center stencil still has the largest EOC and seem to be the best approximation of the analytical solution. Furthermore, the forward stencil variant seem to overestimate the analytical solution while the backward stencil variant seem to underestimate it. We also see that for the smaller values of $\epsilon$ that the forward stencil variant diverges a lot from the analytical solution, while the backward solution continues to match the analytical solution somewhat well. Another interesting aspect to consider is the EOC. We can see that, for smaller $\epsilon$, the EOC is negative for both the forward and backward stencil variant. This tells us that the error increases with more detailed meshes, so we can not reduce the error by increasing the mesh detail.

```python
def adv_diff(f, epsi, stencil):
    N = len(f)-1
    h = 1/N
    A_bis = np.zeros((N+1, N+1))
    A_prime = np.zeros((N+1, N+1))

    for i in range(1, N):
        A_bis[i, i-1] = -1
        A_bis[i, i+1] = -1
        A_bis[i, i] = 2

    if stencil == 'center':
        for i in range(1,N):
            A_prime[i, i+1] = h*0.5*1/epsi
            A_prime[i, i-1] = h*0.5*-1/epsi
    elif stencil == 'forward':
        for i in range(1,N):
            A_prime[i,i] = h*-1/epsi
            A_prime[i,i+1] = h/epsi
    elif stencil == 'backward':
        for i in range(1,N):
            A_prime[i,i] = h*1/epsi
            A_prime[i,i-1] = -h/epsi

    A = A_bis + A_prime
    A[0,0] = 1
    A[N,N] = 1

    F = f*h**2/epsi
    F[0] = 0
    F[N] = 0

    U = la.solve(A, F)
    return U
```

```python
# task a)
epsi = 1
N = 32
x = np.linspace(0,1,N+1)
f = np.ones(x.shape)

U = adv_diff(f,epsi, 'center')

x_analytic = np.linspace(0,1,1001)
U_analytic = x_analytic - (np.exp((x_analytic-1)/epsi) - np.exp(-1/epsi))/(1-np.exp(-1/epsi))

plt.plot(x,U, 'x--r', label='FDM approx.')
plt.plot(x_analytic, U_analytic, '-g', label='Analytical solution')
plt.legend()
plt.title(f"Exc. 5b. epsilon={epsi}, mesh size N = {N}")
plt.show()
```

```python
Ns = [8,16,32,64,128]
markers = ['xr--', 'xg--', 'xb--', 'xk--', 'xc--']
epsilons = [1, 0.1,0.01,0.001]
stencils = ['center', 'forward','backward']
for stencil in stencils:
    print(f"STENCIL: {stencil}")
    for epsi in epsilons:
        errors = []
        for N, marker in zip(Ns, markers):
            x = np.linspace(0,1,N+1)
            f = np.ones(x.shape)

            U = adv_diff(f, epsi, stencil)
            u = x - (np.exp((x-1)/epsi) - np.exp(-1/epsi))/(1-np.exp(-1/epsi))
            errs = np.abs(U-u)
            errors.append(max(errs))

            plt.figure(epsilons.index(epsi))
            plt.plot(x,U, marker, label=f'N={N}')
            plt.title(f'Exc 5c. Epsi={epsi}')

        # analytic
        x = np.linspace(0,1,100)
        U_analytic = x - (np.exp((x-1)/epsi) - np.exp(-1/epsi))/(1-np.exp(-1/epsi))
        plt.plot(x,U_analytic, '-g', label='Analytical solution')
        plt.legend()
        plt.show()

        #do error stuff
        eocs = []
        for i in range(0,len(errors)-1):
            eoc = np.log(errors[i]/errors[i+1])/np.log(2)
            eocs.append(eoc)
        #eocs.append(np.nan)

        data = {"Error":errors[:len(errors)-1], "EOC":eocs}
        df = pd.DataFrame(data, index=Ns[:len(Ns)-1])
        df = df[["Error", "EOC"]]
        df.index.name="N"
        df = df.style.set_caption(f'epsilon={epsi}')
        display(df)
```
